```python
import pandas as pd
```

```python
data = {
    "Order_ID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Laptop","Headphones","Mouse","Keyboard","Monitor","Laptop","Mouse"],
    "Category": ["Electronics","Electronics","Accessories","Accessories","Accessories","Electronics","Electronics","Accessories"],
    "Quantity": [1,2,1,3,2,1,1,4],
    "Unit_Price": [1200,1200,150,25,45,300,1300,25],
    "Region": ["North","South","East","West","North","South","East","West"]
}

df = pd.DataFrame(data)
df
```

```python
df["Revenue"] = df["Quantity"] * df["Unit_Price"]
df
```

```python
df["Revenue"].sum()
```

```python
df.groupby("Product")["Revenue"].sum()
```

```python
df.groupby("Region")["Revenue"].sum()
```

```python
df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
```

```python
df.to_csv("sales_data_project.csv", index=False)
```
